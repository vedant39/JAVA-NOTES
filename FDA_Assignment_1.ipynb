```python
#  Show which the variation of gender among various job titles

df.groupby('Job Title')['Gender'].value_counts().sort_values(ascending=False).unstack()
```

```python
#  Show which Department has most no of Ethinicity people and group them according to Department

df.groupby('Ethnicity')['Job Title'].value_counts().sort_values(ascending=False).unstack()
```

```python
# Show which Department  has most annual salary

df.groupby('Department')['Annual Salary'].mean().sort_values(ascending=False).to_frame()
```

Here we can see that which department has the highest annual salary


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['figure.figsize']=(10,5)
```

```python

df.groupby('Department')['Annual Salary'].mean().sort_values(ascending=False).plot(kind='bar')
plt.xlabel('Department')
plt.ylabel('Average Annual Salary')
plt.title('Average Annual Salary by Department')
plt.show()
```

```python
df.groupby('Job Title')['Annual Salary'].mean().sort_values(ascending=False).plot(kind='bar')
plt.xlabel('Job Title')
plt.ylabel('Average Annual Salary')
plt.title('Average Annual Salary by Job Title')
plt.show()
```

```python
# Histogram of Annual Salary
plt.hist(df['Annual Salary'], bins=10)
plt.xlabel('Annual Salary')
plt.ylabel('Frequency')
plt.title('Histogram of Annual Salary')
plt.show()
```

```python
# Scatter plot of Annual Salary vs. Age
plt.scatter(df['Age'], df['Department'])
plt.xlabel('Age')
plt.ylabel('Annual Salary')
plt.title('Scatter plot of Annual Salary vs. Age')
plt.show()
```

```python
# Pie chart of Employee Distribution by Department
labels = df['Department'].value_counts().index.to_list()
sizes = df['Department'].value_counts().values.tolist()
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie chart of Employee Distribution by Department')
plt.show()
```

```python
# Bar plot of Average Annual Salary by Ethnicity
df.groupby('Ethnicity')['Annual Salary'].mean().plot(kind='bar')
plt.xlabel('Ethnicity')
plt.ylabel('Average Annual Salary')
plt.title('Bar plot of Average Annual Salary by Ethnicity')
plt.show()
```

```python
# Boxplot of Annual Salary by Gender
df.boxplot(column='Annual Salary', by='Gender')
plt.xlabel('Gender')
plt.ylabel('Annual Salary')
plt.title('Boxplot of Annual Salary by Gender')
plt.show()
```
